**Flatten the image** from `(m_train, num_px, num_px, 3)` to `(num_px * num_px * 3, m_train)`

```python
train_set_x_flatten = train_set_x_orig.reshape(m_train, -1).T
test_set_x_flatten = test_set_x_orig.reshape(m_test, -1).T

print('train_set_x_flatten shape:', train_set_x_flatten.shape)
print('train_set_y shape:', train_set_y.shape)
print('test_set_x_flatten shape:', test_set_x_flatten.shape)
print('test_set_y shape:', test_set_y.shape)
```

**Standardize the image**: get the mean and std from all pixels of all images

```python
def normalize_all_pixel(flattened_dataset):
    """
    Normalize the flattened image by mean and std of all pixels in falttened_dataset
    
    Argument:
    flattened_dataset -- a 2-dimensions matrix (num_px * num_px * 3, n_images)
    
    Return:
    The matrix is normalized
    """
    mean = np.sum(flattened_dataset)/float(flattened_dataset.shape[0] * flattened_dataset.shape[1])
    std = np.sqrt(np.sum(np.square(flattened_dataset - mean)) / float(flattened_dataset.shape[0] * flattened_dataset.shape[1]))
    flattened_dataset = (flattened_dataset - mean) / float(std)
    
    return flattened_dataset

train_set_x = normalize_all_pixel(train_set_x_flatten)
test_set_x = normalize_all_pixel(test_set_x_flatten)
```

**Common steps for pre-processing** a new dataset

    * Flatten each image
    * Normalize th flattened image

# 2 - One-layer neural network

- **Very simple neural network**
<img src="images/LogReg_kiank.png">

- **Math** for Inference process for one point $x^{(i)}$:
$$z^{(i)} = w^T x^{(i)} + b \tag{1}$$
$$\hat{y}^{(i)} = a^{(i)} = sigmoid(z^{(i)})\tag{2}$$ 
$$ \mathcal{L}(a^{(i)}, y^{(i)}) =  - y^{(i)}  \log(a^{(i)}) - (1-y^{(i)} )  \log(1-a^{(i)})\tag{3}$$

- And the cost function:
$$ J = \frac{1}{m} \sum_{i=1}^m \mathcal{L}(a^{(i)}, y^{(i)})\tag{6}$$

- And the Back Prop: (The position of `X`, `(A - Y)` may vary based on the way they are defined)
$$ \frac{\partial J}{\partial w} = \frac{1}{m}X(A-Y)^T\tag{7}$$
$$ \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^m (a^{(i)}-y^{(i)})\tag{8}$$

# 4 - Build the functions of the model
* Initialize the parameters
* Compute the inference and its gradients
* Update the parameters using the gradient descent (optimization)
* Use learned (w, b) to predict the labels

## 4.1 Sigmoid function

```python
def sigmoid(z):
    """
    Compute the sigmoid in Logistic Regression
    
    Arguments:
    z: A tensor
    
    Return:
    Sigmoid value
    """
    return 1/(1 + np.exp(-z))
```

```python
print("Sigmoid([0, 10]):", sigmoid(np.array([0, 10])))
```

## 4.2 - Initialize the parameters

```python
def initialize_with_zeros(dim):
    """
    Create the vector zeros w (dim, 1) and scalar b = 0
    
    Argument:
    dim -- the length of vector w
    
    Return
    w -- initialized vector w (dim, 1)
    b -- initialized b = 0
    """
    
    w = np.zeros((dim, 1), dtype=np.float32)
    b = 0
    
    return w, b
```

```python
dim = 2
w, b = initialize_with_zeros(dim)
print('w =', w, 'b =', b)
```

# 4.3 - Forward and its gradient

```python
def forward_gradient(w, b, X, Y):
    """
    Implement the forward propagation and its gradient explained in part 3
    
    Argument:
    w -- weight, size(num_px * num_px * 3, 1)
    b -- bias, scalar
    X -- dataset, size(num_px * num_px * 3, n_images)
    Y -- ground truth, 0: non_cat and 1: cat, size(1, n_images)
    
    Return:
    cost -- negative log-likelihood for logistics regression
    dw -- gradient of w, same shape with w
    db -- gradient of b, same shape with b
    """
    
    # Forward
    A = sigmoid(np.dot(w.T, X) + b)
    cost = -np.mean(Y*np.log(A) + (1-Y)*np.log(1-A))
    
    # Backward
    m = X.shape[1]
    dw = np.dot(X, (A - Y).T)/m
    db = np.mean(A - Y)
    
    grads = {'dw': dw, 'db': db}
    
    return grads, cost
```

```python
w, b, X, Y = np.array([[1.],[2.]]), 2., np.array([[1.,2.,-1.],[3.,4.,-3.2]]), np.array([[1,0,1]])
grads, cost = forward_gradient(w, b, X, Y)
print ("dw = " + str(grads["dw"]))
print ("db = " + str(grads["db"]))
print ("cost = " + str(cost))
```

## 4.4 - Optimization

```python
def optimize(w, b, X, Y, num_iterations, learning_rate, print_every = 10):
    """
    Optimize the parameters w and b by using the gradients
    
    Arguments:
    w -- weight, size(num_px * num_px * 3, 1)
    b -- bias, scalar
    X -- dataset, size(num_px * num_px * 3, n_images)
    Y -- ground truth, 0: non_cat and 1: cat, size(1, n_images)
    num_iterations -- number of loops in the optimization
    learning_rate -- learning rate of gradient descent
    print_every -- number of loops the cost are printed
    
    Returns:
    params: dictinary containing w and b at the end of optimization process
    grads: dictinary containing gradient of w and b (dw and db) at the end of optimization process
    costs: the list of all costs computed during the optimiation. It is used to plot the curve
    """
    costs = []
    for it in range(num_iterations):
        # Forward propagation and its gradient, cost
        grads, cost = forward_gradient(w, b, X, Y)
        
        # Retrieve the derivatives from grads
        dw, db = grads['dw'], grads['db']
        
        # Update the parameters
        w = w - learning_rate*dw
        b = b - learning_rate*db
        
        # Record and print the cost after print_every loops
        if (it + 1)%print_every == 0:
            costs.append(cost)
            print('Cost after', it+1, 'iterations is', cost)
            
    params = {'w': w, 'b': b}
    grads = {'dw': dw, 'db': db}
    
    return params, grads, costs
```

```python
params, grads, costs = optimize(w, b, X, Y, num_iterations=100, learning_rate=.009, print_every=10)

print ("w = " + str(params["w"]))
print ("b = " + str(params["b"]))
print ("dw = " + str(grads["dw"]))
print ("db = " + str(grads["db"]))
```

## 4.5 - Predict
- The 4 steps is the main steps of very basic neural networks, below is `predict` function

```python
def predict(w, b, X, threshold = 0.5):
    """
    Predict the label after the w and b is optimized
    
    Arguments:
    w -- weight, size(num_px * num_px * 3, 1)
    b -- bias, scalar
    X -- dataset, size(num_px * num_px * 3, n_images)
    threshold -- threshold for sigmoid
    
    Returns:
    The probability of 0/1 and predict 0/1
    """
    A = sigmoid(np.dot(w.T, X) + b)
    
    return np.where(A > threshold, A, (1-A)), np.where(A > threshold, 1, 0)
```

```python
w = np.array([[0.1124579],[0.23106775]])
b = -0.3
X = np.array([[1.,-1.1,-3.2],[1.2,2.,0.1]])
print (predict(w, b, X))
```

# 5 - Merge all functions

```python
def model(X_train, Y_train, X_test, Y_test, num_iterations=2000, learning_rate=.5, print_every=100):
    """
    Build the logistic regression by merging all modules
    
    Arguments:
    X_train -- (num_px * num_px * 3, m_train)
    Y_train -- (1, m_train)
    X_test -- (num_px * num_px * 3, m_train)
    Y_test -- (1, m_test)
    num_iterations -- number of loops in the optimization
    learning_rate -- learning rate of gradient descent
    print_every -- number of loops the cost are printed
    
    Returns:
    d - dictionary contains all information about the model
    """
    # Initialize the parameters with zeros
    w, b = initialize_with_zeros(X_train.shape[0])
    
    # Optimization
    params, grads, costs = optimize(w, b, X_train, Y_train, num_iterations, learning_rate, print_every)
    
    # Retrieve the parameters w and b
    w, b = params['w'], params['b']
    
    # Predict the train and test set
    proba_prediction_train, Y_prediction_train = predict(w, b, X_train)
    proba_prediction_test, Y_prediction_test = predict(w, b, X_test)
    
    # Train and Test Error
    print('Train Accuracy:', np.mean(np.equal(Y_prediction_train, Y_train))*100)
    print('Test Accuracy:', np.mean(np.equal(Y_prediction_test, Y_test))*100)
    
    d = {'costs': costs,
        'Y_prediction_train': Y_prediction_train,
        'Y_prediction_test': Y_prediction_test,
        'w': w,
        'b': b,
        'learning_rate':learning_rate,
        'num_iterations': num_iterations}
    
    return d
```

```python
d = model(train_set_x, train_set_y, test_set_x, test_set_y, num_iterations = 2000, learning_rate = 0.005, print_every=200)
```

```python
plt.plot(d['costs'])
plt.xlabel('iterations(per hundreds)')
plt.ylabel('cost')
plt.title('Learning rate = 0.005')
```

# 6 - Further Analysis

```python
learning_rates = [0.01, 0.001, 0.0001]
models = {}

for i in learning_rates:
    print('For learning rate:', i)
    models[str(i)] = model(train_set_x, train_set_y, test_set_x, test_set_y, num_iterations = 1500, learning_rate = i, print_every=300)
    print('\n' + '-----------------------------------------------------' + '\n')
    
for i in learning_rates:
    plt.plot(models[str(i)]['costs'], label=i)
    
plt.xlabel('iterations(per hundreds)')
plt.ylabel('cost')

plt.legend(loc='upper right')
```

**Interpretation**: 
- Different learning rates give different costs and thus different predictions results.
- If the learning rate is too large (0.01), the cost may oscillate up and down. It may even diverge (though in this example, using 0.01 still eventually ends up at a good value for the cost). 
- A lower cost doesn't mean a better model. You have to check if there is possibly overfitting. It happens when the training accuracy is a lot higher than the test accuracy.
