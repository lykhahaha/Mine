# 4 - Neural Network Model
Simple Logistic Regression is not working very well. We will build one-layer neural network
![](images/classification_kiank.png)
The model is quite similar with assignment in Week 2, but add input layer in the beginning

**Math** for Inference process for one point $x^{(i)}$:
$$z^{[1] (i)} =  W^{[1]} x^{(i)} + b^{[1]}\tag{1}$$ 
$$a^{[1] (i)} = \tanh(z^{[1] (i)})\tag{2}$$
$$z^{[2] (i)} = W^{[2]} a^{[1] (i)} + b^{[2]}\tag{3}$$
$$\hat{y}^{(i)} = a^{[2] (i)} = \sigma(z^{ [2] (i)})\tag{4}$$
$$y^{(i)}_{prediction} = \begin{cases} 1 & \mbox{if } a^{[2](i)} > 0.5 \\ 0 & \mbox{otherwise } \end{cases}\tag{5}$$

Given the predictions on all the examples, you can also compute the cost $J$ as follows: 
$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large\left(\small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right)  \large  \right) \small \tag{6}$$

Reminder: Basic steps to build the model
    1. Define neural network structure (number of input units, number of hidden layer, etc)
    2. Initialize the parameters
    3. Loops
        - Implement forward propagation
        - Complute loss
        - Implement back propagation to get the gradients
        - Update the parameters
        
## 4.1 - Define the neural network structure

```python
def layer_size(X, Y):
    """
    Arguments:
    X -- the input (input features size, # of points)
    Y -- the output (output features size, # of points)
    
    Returns:
    n_x -- the size of input layer
    n_y -- the size of output layer
    n_h -- the size of hidden layer
    """
    n_x = X.shape[0]
    n_h = 4
    n_y = Y.shape[0]
    return n_x, n_h, n_y
```

```python
X_assess, Y_assess = layer_sizes_test_case()
n_x, n_h, n_y = layer_size(X_assess, Y_assess)
print('The size of input layer:', n_x)
print('The size of hidden layer:', n_h)
print('The size of output layer:', n_y)
```

## 4.2 - Initialize the model parameters

```python
def initialize_parameters(n_x, n_h, n_y):
    """
    Arguments:
    n_x -- the size of input layer
    n_y -- the size of output layer
    n_h -- the size of hidden layer
    
    Returns:
    params -- dictionary contains:
                W1 -- weihgt matrix (n_h, n_x)
                b1 -- bias vector (n_h, 1)
                W2 -- weight matrix (n_y, n_h)
                b2 -- bias vector (n_y, 1)
    """
    np.random.seed(2)
    
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros((n_h, 1), dtype=np.float32)
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros((n_y, 1), dtype=np.float32)
    
    params = {'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2}
    
    return params
```

```python
n_x, n_h, n_y = initialize_parameters_test_case()

params = initialize_parameters(n_x, n_h, n_y)
print('W1=', params['W1'])
print('b1=', params['b1'])
print('W2=', params['W2'])
print('b2=', params['b2'])
```

## 4.3 The loop

Implement the forward propagation

```python
def forward_propagation(X, params):
    """
    Arguments:
    X -- input matrix (n_x, m -- # of points)
    params -- outputs of initialization function
    
    Return:
    A2 -- output of model
    cache: dictionary contains Z1, A1, Z2 and A2
    """
    # Retieve the paramters
    W1, b1, W2, b2 = params['W1'], params['b1'], params['W2'], params['b2']
    
    # Implement the forward propagation
    Z1 = np.dot(W1, X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    
    assert(A2.shape == (1, X.shape[1]))
    
    cache = {'Z1': Z1, 'A1': A1, 'Z2': Z2, 'A2': A2}
    
    return A2, cache
```

```python
X_assess, params = forward_propagation_test_case()

A2, cache = forward_propagation(X_assess, params)

print(np.mean(cache['Z1']), np.mean(cache['A1']), np.mean(cache['Z2']), np.mean(A2))
```

```python
def compute_cost(A2, Y):
    """
    Compute the negative log-likelihood cost
    Arguments:
    A2 -- the output of forward propagation function
    Y -- the ground truth
    Returns:
    Return the cost value
    """
    return -np.mean(Y*np.log(A2) + (1-Y)*np.log(1-A2))
```

```python
A2, Y_assess, params = compute_cost_test_case()
print('Cost:', compute_cost(A2, Y_assess))
```

Implement the backward propagation
![](images/grad_summary.png)

To compute dZ1 you'll need to compute $g^{[1]'}(Z^{[1]})$. Since $g^{[1]}(.)$ is the tanh activation function, if $a = g^{[1]}(z)$ then $g^{[1]'}(z) = 1-a^2$. So you can compute 
    $g^{[1]'}(Z^{[1]})$ using `(1 - np.power(A1, 2))`.

```python
def backward_propagation(cache, params, X, Y):
    """
    Arguments:
    cache -- output of forward propagation
    params -- output of initialize parameters
    X -- the input (input features size, # of points)
    Y -- the output (output features size, # of points)
    
    Returns:
    grads -- dictionary contains the gradients with respect to the parameters
    """
    m = X.shape[1] # number of points
    # Retrieve A1 and A2 from cache
    A1, A2 = cache['A1'], cache['A2']
    
    # Retrieve weight and bias from params
    W1, b1, W2, b2 = params['W1'], params['b1'], params['W2'], params['b2']
    
    # Backward propagation process
    E2 = A2 - Y # shape (output features size, m)
    dW2 = 1/m * np.dot(E2, A1.T) # shape (output features size, # of hidden layer)
    db2 = 1/m * np.sum(E2, axis=1, keepdims=True) # shape (output features size, 1)
    
    E1 = np.dot(W2.T, E2) * (1 - A1**2) # shape (# of hidden layer, m)
    dW1 = 1/m * np.dot(E1, X.T) # shape (# of hidden layer, input features size)
    db1 = 1/m * np.sum(E1, axis=1, keepdims=True) # shape (# of hidden layer, 1)
    
    grads = {'dW1': dW1, 'db1': db1, 'dW2': dW2, 'db2': db2}
    
    return grads
```

```python
parameters, cache, X_assess, Y_assess = backward_propagation_test_case()

grads = backward_propagation(cache, parameters, X_assess, Y_assess)
print ("dW1 = "+ str(grads["dW1"]))
print ("db1 = "+ str(grads["db1"]))
print ("dW2 = "+ str(grads["dW2"]))
print ("db2 = "+ str(grads["db2"]))
```

```python
def update_parameters(parameters, grads, learning_rate=1.2):
    """
    Arguments:
    parameters -- output of initialize parameters
    grads -- output of backward propagation
    
    Returns:
    params -- dictionary contains updated parameters
    """
    # Retrieve weight and bias from parameters
    W1, b1, W2, b2 = parameters['W1'], parameters['b1'], parameters['W2'], parameters['b2']
    
    # Retrieve gradient wrt parameters
    dW1, db1, dW2, db2 = grads['dW1'], grads['db1'], grads['dW2'], grads['db2']
    
    # Update the parameters
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    params = {'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2}
    
    return params
```

```python
parameters, grads = update_parameters_test_case()
parameters = update_parameters(parameters, grads)

print("W1 = " + str(parameters["W1"]))
print("b1 = " + str(parameters["b1"]))
print("W2 = " + str(parameters["W2"]))
print("b2 = " + str(parameters["b2"]))
```

# 4.4. Merge all functions in 4.1, 4.2 and 4.3

```python
def nn_model(X, Y, n_h, num_iterations=10000, print_every=100, learning_rate=1.2):
    """
    Arguments:
    X -- the input (input features size, # of points)
    Y -- the output (output features size, # of points)
    n_h -- number of hidden layer
    num_iteration -- number of iterations of gradient descent loop
    print_every -- number of iterations cost will be printed
    
    Returns:
    params -- all paraemters learnt in gradient descent loop
    """
    n_x = layer_size(X, Y)[0]
    n_y = layer_size(X, Y)[2]
    
    # Initialize parameters
    parameters = initialize_parameters(n_x, n_h, n_y)
    W1, b1, W2, b2 = parameters['W1'], parameters['b1'], parameters['W2'], parameters['b2']
    
    # Loop
    for i in range(num_iterations):
        # Forward propagation
        A2, cache = forward_propagation(X, parameters)
        
        # Cost function
        cost = compute_cost(A2, Y)
        
        # Backward propagation
        grads = backward_propagation(cache, parameters, X, Y)
        
        # Update parameter by gradient descent
        params = update_parameters(parameters, grads, learning_rate)
        
        if print_every != None and (i+1)%print_every==0:
            print('Cost after', i+1, 'is', cost)
            
    return params
```

```python
X_assess, Y_assess = nn_model_test_case()

parameters = nn_model(X_assess, Y_assess, 4, num_iterations=10000, print_every=500)
print("W1 = " + str(parameters["W1"]))
print("b1 = " + str(parameters["b1"]))
print("W2 = " + str(parameters["W2"]))
print("b2 = " + str(parameters["b2"]))
```

# 4.5. Predictions

```python
def predict(parameters, X, threshold=0.5):
    """
    Using the learned parameters, predicts a class for each example in X
    
    Arguments:
    parameters -- output of nn_model
    X -- input data of size (input features size, m)
    
    Returns
    predictions -- vector of predictions of our model (red: 0 / blue: 1)
    """
    
    # Forward propagation with learnt paramters
    A2, cache = forward_propagation(X, parameters)
    
    return np.where(A2 > threshold, 1, 0)
```

# 4.6. Tuning hidden layer size

```python
plt.figure(figsize=(16, 32))
hidden_layer_size = [1, 2, 3, 4, 5, 20, 50]
for i, n_h in enumerate(hidden_layer_size):
    plt.subplot(5, 2, i+1)
    plt.title('Hidden layer size %d' % n_h)
    parameters = nn_model(X, Y, n_h, num_iterations=5000, print_every=None)
    plot_decision_boundary(lambda x: predict(parameters, x.T), X, Y)
    predictions = predict(parameters, X)
    print('Train Accuracy for', n_h, 'is', np.mean(np.equal(predictions, Y))*100)
```

**Intepretation:**

* The larger models (with more hidden units) are able to fit the training set better, until eventually the largest models overfit the data.
* The best hidden layer size seems to be around n_h = 5. Indeed, a value around here seems to fits the data well without also incurring noticable overfitting.

# 5. Test on other datasets

```python
noisy_circles, noisy_moons, blobs, gaussian_quantiles, no_structure = load_extra_datasets()
datasets = {"noisy_circles": noisy_circles,
            "noisy_moons": noisy_moons,
            "blobs": blobs,
            "gaussian_quantiles": gaussian_quantiles}
```

```python
def test_dateset(dataset_name):
    dataset = dataset_name
    X, Y = datasets[dataset]
    X, Y = X.T, Y.reshape(1, -1)
    
    if dataset_name == 'blobs':
        Y = Y % 2
        
    plt.figure(figsize=(16, 32))
    plt.subplot(5, 2, 1)
    plt.title('Original dataset %s plot' % dataset_name)
    plt.scatter(X[0, :], X[1, :], c=Y, s=40, cmap=plt.cm.Spectral)
    
    
    hidden_layer_size = [1, 2, 3, 4, 5, 20, 50]
    for i, n_h in enumerate(hidden_layer_size):
        plt.subplot(5, 2, i+2)
        plt.title('Hidden layer size %d' % n_h)
        parameters = nn_model(X, Y, n_h, num_iterations=5000, print_every=None)
        plot_decision_boundary(lambda x: predict(parameters, x.T), X, Y)
        predictions = predict(parameters, X)
        print('Train Accuracy for', n_h, 'is', np.mean(np.equal(predictions, Y))*100)
        
test_dateset('noisy_moons')
```
