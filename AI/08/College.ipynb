* Now I use tensorflow to implement softmax regreession for IRIS dataset

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from sklearn.metrics import confusion_matrix
```

```python
all_data = pd.read_csv('iris.csv', names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])
all_data.head()
```

```python
scaler = StandardScaler()
all_data[all_data.columns[:-1]] = scaler.fit_transform(all_data[all_data.columns[:-1]])
pd.concat
all_data.head()
```

```python
all_data = pd.concat([all_data, pd.Series(np.ones(len(all_data)))], axis=1)
all_data.head()
```

```python
all_data[all_data.select_dtypes(include='object').columns].head()
```

```python
train_x, test_x, train_y, test_y = train_test_split(all_data[all_data.select_dtypes(exclude='object').columns], all_data[all_data.select_dtypes(include='object').columns], test_size = 0.3, random_state = 1)
```

```python
train_y = pd.get_dummies(train_y)
test_y = pd.get_dummies(test_y)
```

```python
x = tf.placeholder(tf.float64, name='x')
y = tf.placeholder(tf.float64, name='y')
```

```python
w_shape = (train_x.shape[1], 3)
w = tf.Variable(np.random.normal(0, np.sqrt(2.0/np.sum(w_shape)), w_shape), name='weights')
```

```python
numerator = tf.exp(tf.matmul(x, w))
denominator = tf.reduce_sum(numerator, 1, keepdims=True)
pred = tf.divide(numerator, denominator)
```

```python
num_epoch = 10000
learning_rate = 0.01
cost = -tf.reduce_mean(y * tf.log(pred))
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
all_test_loss = []
```

```python
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for e in range(num_epoch):
        sess.run(cost, feed_dict = {x: train_x, y: train_y})
        test_loss = sess.run(cost, feed_dict = {x: test_x, y: test_y})

        sess.run(optimizer, feed_dict = {x: train_x, y: train_y})
        if all_test_loss and test_loss - all_test_loss[-1] > 0:
            break
        
        all_test_loss.append(test_loss)
        if e % 1000 == 0:
            print("Epoch %d: test loss: %.5f" % (e+1, test_loss))
    y_hat = sess.run(pred, feed_dict={x: test_x})
```

```python
test_y = np.argmax(test_y.values, axis=1)
y_hat = np.argmax(y_hat, axis=1)
```

```python
confusion_matrix(test_y, y_hat)/np.sum(confusion_matrix(test_y, y_hat), axis=0)
```

```python
confusion = confusion_matrix(test_y, y_hat)/np.sum(confusion_matrix(test_y, y_hat), axis=0)
confusion.flatten()[0::4]
```
